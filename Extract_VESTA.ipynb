```python
import pandas as pd

columns = ["No.", "mat", "R_atomic(A)", "R_ionic(A)", "R_Waals(A)", "R", "G", "B"]
df_elm = pd.DataFrame(columns=columns)

line = " "
with open("elements.ini", "r") as f:
    idx = 0
    while line:
        line = f.readline()
        if not line:
            break
            
        data = line.rstrip("\n").rstrip("\r").lstrip(" ").split()
        df_elm.loc[idx] = data
        idx += 1
    
df_elm["No."] = df_elm["No."].astype("int")
df_elm.iloc[:, 2:] = df_elm.iloc[:, 2:].astype("float")
df_elm.to_csv("VESTA_elements.csv", index=False)        
```

```python
mat_annot = {}

line = " "
with open("materials_list.txt", "r") as f:
    idx = 0
    while line:
        line = f.readline()
        if not line:
            break
            
        data = line.rstrip("\n").rstrip("\r").lstrip(" ").replace("-", "").split()
        print data[0], data[1]
        mat_annot.update({data[0]: data[1]})
        idx += 1

mat_annot
```

```python
with open("paraview_VESTA_atom.json", "w") as j:
    j.write('[\n\t{\n\t\t"Annotations" :\n\t\t[\n')
    for i in range(df_elm.shape[0]):
        mat = df_elm["mat"].loc[i]
        if mat == "D":
            annot = "Deuterium"
        elif mat == "XX":
            annot = "etc."
        else:
            annot = mat_annot[mat]
            
        j.write('\t\t\t"%s",\n\t\t\t"%s",\n' %(mat, annot))
    
    j.write('\t\t],\n\t\t"IndexedColors" :\n\t\t[\n')
    for i in range(df_elm.shape[0]):
        j.write('\t\t\t%s,\n' %(df_elm["R"].loc[i]))
        j.write('\t\t\t%s,\n' %(df_elm["G"].loc[i]))
        j.write('\t\t\t%s,\n' %(df_elm["B"].loc[i]))
        
    j.write('\t\t],\n\t\t"Name" : "VESTA atom"\n\t}\n]')

    
```
